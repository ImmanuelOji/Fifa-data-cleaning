```python
## Converting the values in inches to cm
import re
```

```python
def convert_to_cm(height):
    """Converts a height string from feet and inches to centimeters"""
    match = re.match(r"(\d+)'(\d+)\"", height)
    if match:
        feet = int(match.group(1))
        inches = int(match.group(2))
        return round(feet * 30.48 + inches * 2.54)
    else:
        return height
```

```python
# apply conversion function to Height column
fifa['Height'] = fifa['Height'].apply(convert_to_cm)
```

```python
## checking to see if everything is in cm
```

```python
fifa.Height.unique()
```

## converting from lbs to kg

```python
def convert_to_kg(weight):
    """Converts a weight string from lbs to kg"""
    match = re.match(r"(\d+)(lbs|kg)", weight)
    if match:
        value = int(match.group(1))
        unit = match.group(2)
        if unit == 'lbs':
            return round(value * 0.45359237)
        else:
            return value
    else:
        return weight
```

```python
# apply conversion function to Weight column
fifa['Weight'] = fifa['Weight'].apply(convert_to_kg)
```
